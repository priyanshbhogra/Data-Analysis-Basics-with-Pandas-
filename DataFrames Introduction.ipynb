DROP ROWS WITH NULL VALUES

```python
nba_roster2 = pd.read_csv('pandas/nba.csv')

nba_roster2.head()
```

```python
nba_roster2 = nba_roster2.dropna()#drop all the rows which have any null value

nba_roster2.head(5)
#as you can see below, row with John Holland deleted; last row with null values also dropped
#from 458 rows to 364 rows
```

```python
nba_roster.dropna(axis= 1, how = "all") #how --> delete rows where all values are Null values 
#axis --> only delete from columns

nba_roster.dropna(subset="Salary") #delete rows where Salary has null values
```

FILL IN NULL VALUES WITH THE .fillna() METHOD

```python
nba_roster3 = pd.read_csv("pandas/nba.csv")
nba_roster3.head(5)
```

```python
nba_roster3.fillna(0) #will replace all the null values with 0 which might not be feasible for some columns
```

```python
nba_roster3["Salary"].fillna(0, inplace=True) #only replace the salary column's null values with 0
```

```python
nba_roster3.head(5)
```

```python
nba_roster3["College"].fillna("Community College/Junior College", inplace=True)
nba_roster3.head(5)
```

CONVERT DATAFRAME COLUMN TYPES WITH THE .astype() METHOD

```python
nba_roster4 = pd.read_csv("pandas/nba.csv").dropna(how="all")
nba_roster4["Salary"].fillna(0, inplace=True)
nba_roster4["College"].fillna("None", inplace=True)
nba_roster4.head(6)
```

```python
nba_roster4.info() #gives info if there are any null values, how many float columns
```

```python
nba_roster4["Age"] = nba_roster4["Age"].astype("int")
```

```python
nba_roster4.info()
```

```python
nba_roster4["Number"] = nba_roster4["Number"].astype("int")
```

```python
nba_roster4["Salary"] = nba_roster4["Salary"].astype("int")
```

```python
nba_roster4["Position"] = nba_roster4["Position"].astype("category")
```

```python
nba_roster4["Team"] = nba_roster4["Team"].astype("category")
```

SORTING A DATAFRAME WITH .sort_values() METHOD 

```python
#FIND THE OLDEST PLAYER IN THE ROSTER

nba_roster4.sort_values("Age", ascending = False).head(1)
```

```python
nba_roster4.sort_values(["Team", "Name"], ascending=[True, False]).head() #sort by multiple columns 
```

SORT DATA FRAME WITH THE .sort_index() METHOD

```python
nba_roster4.sort_index(ascending=True)
```

RANK VALUES WITH THE .rank() METHOD

```python
#RANKING THE SALARIES IN NBA (nba_roster4)

nba_roster4["Salary Rank"] = nba_roster4["Salary"].rank(ascending=False).astype("int")
```

```python
nba_roster4.head()
```

```python
#FIND THE HIGHEST SALARY AND RANK

nba_roster4.sort_values("Salary", ascending=False).head(5)
```
